```python
import numpy as np
import matplotlib.pyplot as plt

molemmat=False
euler=True

y_0=2
x_lim=300
h=0.1

b=0.1
m=.0

c=0.0009

H=0.01

t_rauhoitus=2000

x=np.linspace(0,x_lim,int(x_lim/h)+1)
y_euler=np.zeros(len(x))
y_rk=np.zeros(len(x))

y_euler[0]=y_0
y_rk[0]=y_0

def f(t,y):
    if t<t_rauhoitus:
        return b*y-m*y-c*y**2-H
    else:
        return b*y-m*y-c*y**2

for i in range(1,len(y_euler)):
    y_euler[i]=y_euler[i-1]+h*f(h*(i-1),y_euler[i-1])
    
    k_1=f(h*(i-1),y_rk[i-1])
    k_2=f(h*(i-1)+h/2,y_rk[i-1]+k_1*h/2)
    k_3=f(h*(i-1)+h/2,y_rk[i-1]+k_2*h/2)
    k_4=f(h*(i-1),y_rk[i-1]+k_3*h)
    y_rk[i]=y_rk[i-1]+(k_1+2*k_2+2*k_3+k_4)*h/6



plt.axis([0,x_lim,0,150])

if molemmat:
    plt.plot(x,y_euler,x,y_rk)
elif euler:
    plt.plot(x,y_euler)
else:
    plt.plot(x,y_rk)
```
